# Efficiency from Absorption Profile

import necessary packagaes

```python
%config InlineBackend.figure_format = 'retina' #to make inline images higher res

from math import pi
from math import gamma
from scipy.integrate import trapz
from scipy.interpolate import interp1d
import scipy.io

import numpy as np
import matplotlib.pyplot as plt

import time
```

define plot style

```python
plt.style.use(['presplot'])
```

initialize constants, load data

```python
#constants
c=2.998e8 #m/s
h=4.136e-15 #eV s
k=8.617e-5 #eV/K
q=1.602e-19 #C
t_c=300
t_s=5800
omega_s=6.8e-5
omega_c=pi

#bandgaps
e_step=0.02 #step size for bandgaps 
egs=np.arange(e_step,3.0,e_step).round(3) #bandgaps in eV
egmax=egs[np.abs((egs-1.337).round(3))==min(np.abs(egs-1.337).round(3))]

#energies 
e=np.arange(e_step,8,e_step) #energies in eV

#am1.5 data
am1p5_raw=np.loadtxt("am1p5_raw.txt")

#Urbach stuff
urb=2*k*t_c
theta=1
l=2 #really alpha0*l
xpr=np.array(np.linspace(-25,25,1000),dtype='complex')
```

### Max Efficiency

calculate maximum efficiency given this absorption profile

```python
#incident power
ev=1240/am1p5_raw[::-1,0]
am1p5=1/ev**2*1240*am1p5_raw[::-1,1] #W/(m^2 eV)
pinc_am1p5=trapz(am1p5,ev)

#create interpolant
g=interp1d(ev,am1p5,fill_value="extrapolate")

#initialize efficiency
eff_am1p5=np.zeros(len(egs))
```

run over bandgaps, find max efficiency

```python
ind=0 #initialize index

for eg in egs:
    
    #define voltages
    vs=np.arange(0,eg,e_step) #voltages from from 0 to bandgap
    
    #calculate absorption
    xs=(e-eg)/(urb)
    gs=np.zeros(np.size(xs))
    i=0
    for x in xs:
        integrand=np.exp(-np.abs(xpr))*np.sqrt(x-xpr)
        gs[i]=np.real(1/2*trapz(integrand,xpr)) #gamma(2)=1
        i+=1

    alpha=(urb/(k*t_c))**0.5*gs
    a=1-np.exp(-2*alpha*l)
    
    #calculate light emitted from cell at given voltage
    j0=np.zeros(np.size(vs))
    jsc=np.zeros(np.size(vs))
    index=0
    for v in vs:
        #absorbed from sun
        fermi=np.tanh((e-v)/(4*k*t_c))
        rad_s=g(e)/(q*e) #bb radiation from sun, photons/(m^2*s)
        jsc[index]=q*trapz(a*rad_s*fermi,e) #C/(m^2*s) = A/m^2
        
        #emitted
        s_PL=2*pi*a/(h**3*c**2)*e**2/(np.exp((e-v)/(2*k*t_c))+1)**2
        j0[index]=q*trapz(s_PL,e) #C/(m^2*s) = A/m^2
        index+=1
    
    j=jsc-j0

    pmax=np.max(j*vs)
    eff_am1p5[ind]=pmax/pinc_am1p5
    if eg==egmax: #save I-V data for best device
        jmax=j
        vmax=vs
    
    ind+=1 #index
```

find voc, jsc, etc.

```python
temp=vmax[np.absolute(jmax)==np.min(np.absolute(jmax))]
voc=temp[0]
temp=jmax[np.absolute(vmax)==np.min(np.absolute(vmax))]
jsc=round(temp[0],3)
pce=np.max(jmax*vmax)/pinc_am1p5
ff=np.max(jmax*vmax)/(voc*jsc)
```

### Plots 

plot S-Q efficiency limit

```python
fig,ax=plt.subplots()
ax.plot(egs,eff_am1p5,label='AM1.5')
ax.set_xlabel("Bandgap (eV)")
ax.set_ylabel("Max. Efficiency (%)")
ax.legend(loc="upper right")
ax.set_title('Detailed Balance Efficiency Limit')
ax.set_xlim([0.5,3])
```

plot I-V curve for best solar cell

```python
fig,ax=plt.subplots()
ax.plot(vmax,jmax/10)
ax.set_xlabel("Voltage (V)")
ax.set_ylabel("Current Density ($mA/cm^2$)")
ax.set_title('Best IV Curve')
ax.set_xlim([0.1,voc*1.3]);
ax.set_ylim([0,40]);

ivdata = '$V_{oc}$='+str(voc*1000)+' mV\n$J_{sc}=$'+str(round(jsc/10.0,1))+' $mA/cm^2$'\
    '\nFF='+ str(round(ff,3))+'\nPCE='+str(round(pce*100,1))+'%'
ax.text(0.2,10,ivdata);
```

### Absorption

define absorption profile

```python
#Katahara and Hillhouse - convolve band-to-band with Urbach 
eg=egmax
xs=np.array((e-eg)/(urb),dtype='complex')
xpr=np.array(np.linspace(-20,20,500),dtype='complex')
gs=np.zeros(np.size(xs))
test=np.zeros(np.size(xs),dtype='complex')
i=0
for x in xs:
    #integrand=np.exp(-np.abs(xpr)**theta)*np.sqrt(x-xpr)
    integrand=np.exp(-np.abs(xpr))*np.sqrt(x-xpr)
    #gs[i]=np.real(1/(2*gamma(1+1/theta))*trapz(integrand,xpr))
    gs[i]=np.real(1/2*trapz(integrand,xpr))
    test[i]=1/2*trapz(integrand,xpr)
    i+=1

alpha=(urb/(k*t_c))**0.5*gs
a=1-np.exp(-2*alpha*l)
```

plot

```python
fig,ax=plt.subplots()
ax.plot(e,a,label='$\gamma=$'+str(round(urb/(k*t_c),1)))
ax.set_xlabel("Energy (eV)")
ax.set_ylabel("A")
ax.set_title('Absorption')
ax.set_xlim([0.5,2.5])
ax.set_ylim([0,1])
ax.axvline(x=eg,color='grey')
ax.legend()
```

### Photoluminescence

calculate PL spectrum using Generalized Planck's Law

```python
qfls=voc
s_PL=2*pi*a/(h**3*c**2)*e**2/(np.exp((e-qfls)/(2*k*t_c))+1)**2 #bb radiation from cell, photons/(m^2*s)
```

plot

```python
fig,ax=plt.subplots(figsize=(6,2))
ax.plot(e,s_PL,label='$\gamma=$'+str(round(urb/(k*t_c),1)))
ax.set_xlabel("Energy (eV)")
ax.set_ylabel("PL")
ax.set_title('Photoluminescence')
ax.legend()
ax.set_xlim([0,2.5])
```
