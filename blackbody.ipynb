constants

```python
#constants
c=2.998e8 #m/s
h=6.626e-34 #J s
hbar=h/(2*np.pi);
k=1.381e-23 #J/K
q=1.602e-19 #C

t_0=300
omega_0=pi

t_s=5800
omega_s=6.8e-5
```

### BB spectrum in nm

```python
#solar irradiation assuming 5800 K blackbody
t=t_s
lda_step=1
lda=np.arange(10,10000,lda_step)*1e-9
bblda=2*h*c**2/lda**5*1/(np.exp(h*c/(lda*k*t))-1)*omega_s

total=np.trapz(bblda,lda)
total
```

### BB spectrum in eV

```python
#plot an arbitrary blackbody spectrum in eV
t=t_s
omega=omega_s

e_step=0.001
e=np.arange(e_step,10,e_step)*q

#bb=2*q*omega/(h**3*c**2)*e**2/(np.exp(e/(k*t))-1) #photons/m^2/s
bbe=q*2*omega/(h**3*c**2)*e**3/(np.exp(e/(k*t))-1) #multiply by q for 1/J to 1/eV

fig,ax=plt.subplots()

illuminant = illuminants.get_blackbody_illuminant(t)
color=blackbody.blackbody_color(t)
irgb=colormodels.irgb_from_xyz(color)
irgb2=colormodels.irgb_string_from_irgb(irgb)

ax.plot(e/q,bbe,color='black')
ax.set_xlabel("Energy (eV)")
ax.set_ylabel("Spectral Irradiance (W $m^{-2}$ $eV^{-1}$)")
ax.set_xlim([0,8])
ax.set_facecolor(irgb2)
```

### k_rad  
calculate Rrad given a step function absorption

```python
bbc=2*omega_0/(h**3*c**2)*e**2/(np.exp(e/(k*t_0))-1) #photons/m^2/s/J
bbs=2*omega_s/(h**3*c**2)*e**2/(np.exp(e/(k*t_s))-1)

eg=1.34*q

m0=9.109e-31; #kg
mh=1*m0
me=1*m0

Nc=2*(me*k*t_0/(2*np.pi*hbar**2))**(3/2)*1/(100**3) #1/cm^3
Nv=2*(mh*k*t_0/(2*np.pi*hbar**2))**(3/2)*1/(100**3) #1/cm^3
ni=(Nc*Nv)**0.5*np.exp(-eg/(2*k*t_0))

#assume total absorption above the bandgap
integrand=bbc #units are photons/m^2/s, per energy slice
integrand[e<eg]=0
r=trapz(integrand,e) #photons/m^2/s
r*=1e-4/1e-3 #1/m^2 to 1/cm^2, then divide by thickness in cm (10 um, compare to Wurfel)
br=r/ni**2 #cm^3/s
tau=1/(2*ni*br)

#generation rate
integrand=bbs #units are photons/m^2/s, per energy slice
integrand[e<eg]=0
g=trapz(integrand,e) #photons/m^2/s
g_cm2=g*1e-4
g_cm3=g_cm2/10e-4 #1/m^2 to 1/cm^2, then divide by thickness in cm

#max QFLS
qfls=(eg-k*t_0*np.log(Nc*Nv*br/g_cm3))/q
```
