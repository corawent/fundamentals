# Slab Waveguides

import packages

```python
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
from math import pi
from scipy.optimize import root
```

define plot style

```python
plt.style.use(['presplot'])
```

constants

```python
c=2.998e8 #m/s
mu0=1.257e-6 #SI units
eps0=8.85e-12
```

### input parameters here

```python
d=1 #m
n1=4.5 #waveguide index
n2=1.4 #cladding index
m=6 #max mode index
```

calculate modes

```python
eps1=n1**2*eps0
eps2=n2**2*eps0

kx_0=1/d
kx=kx_0*np.linspace(0.001,m*pi/2,10000)

omega_0=c/d
#omegas=np.linspace(0.0001,10,1000)*omega_0
omegas=np.logspace(-3,1,1000)*omega_0

y1=np.tan(kx*d)
y2=-1/np.tan(kx*d)
y1[y1<0]=0
y2[y2<0]=0
y=y1+y2

all_kxs=np.zeros([np.size(omegas),m])
j=0
for omega in omegas:
    inside=mu0*omega**2*(eps1-eps2)*d**2/(kx*d)**2-1
    inside[inside<0]=np.inf
    y3=np.sqrt(inside)

    kxs=[]
    for i in np.arange(1,m,1):
        diff=abs(y3-y)
        diff_window=diff[np.logical_and(kx*d>((i-1)*pi/2),kx*d<(i*pi/2))]
        kx[diff==min(diff_window)]
    
        if min(diff_window)!=np.inf:
            kxs.append(kx[diff==min(diff_window)])

    all_kxs[j,0:np.size(kxs)]=kxs
    j+=1
```

plot solutions for one omega

```python
#define omega to plot
omega_plot=2*omega_0
inside=mu0*omega_plot**2*(eps1-eps2)*d**2/(kx*d)**2-1
inside[inside<0]=np.inf
y3=np.sqrt(inside)

#plot it
fig,ax=plt.subplots()
ax.plot(kx*d,y,'.')
ax.plot(kx*d,y3,'.')
ax.plot(kx*d,abs(y3-y))
ax.set_ylim([0,10])
ax.set_xlabel('$k_x*d$')
ax.set_ylabel('y')
```

calculate full dispersion relation in kz (kz is along waveguide)

```python
kzs=np.zeros([np.size(omegas),m])
for i in np.arange(0,m,1):
    kzs[:,i]=np.sqrt(omegas**2*mu0*eps1-all_kxs[:,i]**2)
    kzs[:,i][all_kxs[:,i]==0]=np.inf
    
#limiting lines
kzmax=omegas*(mu0*eps1)**0.5
kzmin=omegas*(mu0*eps2)**0.5
```

plot full dispersion relation

```python
fig,ax=plt.subplots()
ax.plot(omegas/omega_0,kzmax/kx_0,'--k')
ax.plot(omegas/omega_0,kzmin/kx_0,'--k')
for i in np.arange(0,m,1):
    ax.plot(omegas/omega_0,kzs[:,i]/kx_0,label='m='+str(i))
ax.set_xlim([0,2])
ax.set_ylim([0,5])
ax.set_xlabel("$\omega$ ($c/d$)")
ax.set_ylabel("$k_z$ (1/d)")
#ax.legend()
```

calculate cutoff frequency for a given mode and thickness. cutoff frequency is minimum frequency for a given mode, so cutoff wavelength is maximum wavelength (lowest energy)

```python
d=1e-9
n=2
omega_c=(n-1)*pi/(2*d)*1/np.sqrt(mu0*(eps1-eps2))
#omega_c*1e-9 #GHz
lambda_c=2*pi*c/omega_c #m
lambda_c 

omega0=c/d
lambda0=2*pi*c/omega0
lambda0

lda=615e-9
omga=2*pi*c/lda
omga*d/c
```

so omega=1 on the plot for a 1 nm waveguide corresponds to a wavelength of 6 nm. 615 nm corresponds to omega=0.01 AKA very low frequency, only TE1 mode accessible. Wavevector and frequency corespond to low-frequency limit (kz=omega c sqrt(eps_cladding))
