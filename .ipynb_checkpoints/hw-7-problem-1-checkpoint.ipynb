# Optical Properties of Solids

Some different models to describe the absorption of light in solids due to free carrier absorption, dielectric resonances, and interband transitions.

import necessary packagaes

```python
%config InlineBackend.figure_format = 'retina' #to make inline images higher res

from math import pi
#from scipy.integrate import trapz
#from scipy.interpolate import interp1d
#import scipy.io

import numpy as np
import matplotlib.pyplot as plt
```

define plot style

```python
plt.style.use(['presplot'])
```

constants

```python
c=2.998e8
hbar=1e-34
me=9.1e-31
eps_0=8.85e-12
e=1.6e-19 #C
```

### Temperature Dependence of Optical Absorption: Free Carriers in Metals


```python
#parameters
n=1e23*1e6 #1/m^3
tau0=1e-14 #s, from A&M pg 9
eps_core=1

#temperatures
Ts=[1/10,1,10]
alphas={}
omega_ps={}

for T in Ts:
    tau=tau0/(T)**0.5

    #calculate dielectric constant
    ev=np.logspace(-3,1.5,500)
    omega=ev*e/hbar
    sigma0=n*e**2*tau/me
    sigma=sigma0/(1-1j*omega*tau)
    eps=(eps_core+1j*sigma/(eps_0*omega))

    #calculate n and k
    nc=eps**(0.5)

    #calculate alpha
    alpha=2*omega*np.imag(nc)/c*1/100

    #calculate plasma frequency
    omega_p=(n*e**2/(me*eps_core*eps_0))**0.5
    ev_p=omega_p*hbar/e
    
    alphas[str(T)]=alpha
    omega_ps[str(T)]=omega_ps
```

plot optical constants

```python
fig,ax=plt.subplots()

ax.axvline(x=ev_p,color="black",linestyle='--',label='$\omega_p$')

for T in Ts:
    ax.plot(ev,alphas[str(T)],label=r'$\alpha_{T='+str(T)+'}$')
    
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Energy (eV)')
ax.set_ylabel('Absorption coefficient $(cm^{-1})$')
ax.set_title(r'$\alpha$(T), Free Carriers in Metals')
ax.legend()
```

Free carrier absorption decreases with increasing temperature in metals because the scattering time for electron-phonon coupling goes as 1/T. 

### Temperature Dependence of Optical Absorption: Free Carriers in Semiconductors


```python
#parameters
m=0.01*me
eps_core=1
exp_fac=np.exp(-0.5/0.025) #assuming a bandgap of roughly an eV
n0=1e21/exp_fac #1/m^3 = 1e15 1/cm^3
tau0=1e-14 #s, from A&M pg 9

#temperatures
Ts=[0.8,1,1.2]
alphas={}
ev_ps={}
ns={}

for T in Ts:
    tau=tau0/(T)**0.5
    n=n0*exp_fac**(1/T) 

    #calculate dielectric constant
    ev=np.logspace(-2,1,500)
    omega=ev*e/hbar
    sigma0=n*e**2*tau/m
    sigma=sigma0/(1-1j*omega*tau)
    eps=(eps_core+1j*sigma/(eps_0*omega))

    #calculate n and k
    nc=eps**(0.5)

    #calculate alpha
    alpha=2*omega*np.imag(nc)/c*1/100

    #calculate plasma frequency
    omega_p=(n*e**2/(m*eps_core*eps_0))**0.5
    ev_p=omega_p*hbar/e
    
    alphas[str(T)]=alpha
    ev_ps[str(T)]=ev_p
    ns[str(T)]=n
```

plot optical constants

```python
fig,ax=plt.subplots()

for T in Ts:
    #ax.axvline(x=ev_p,color="black",linestyle='--',label='$\omega_p$')
    ax.plot(ev,alphas[str(T)],label=r'$\alpha_{T='+str(T)+'}$')
    
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Energy (eV)')
ax.set_ylabel('Absorption coefficient $(cm^{-1})$')
ax.set_title(r'$\alpha$(T), Free Carriers in SCs')
ax.legend()
```

Note the different y-axis scale and different temperatures used from metals to SCs. Free carrier absorption increases with increasing temperature in metals because the carrier concentration increases exponentially as temperature increases.
