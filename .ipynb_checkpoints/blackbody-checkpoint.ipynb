# Blackbody Spectrum

import packages

```python
%config InlineBackend.figure_format = 'retina' #to make inline images higher res

from math import pi
from scipy.integrate import trapz
from scipy.interpolate import interp1d
import scipy.io

import numpy as np
import matplotlib.pyplot as plt
```

set plot format

```python
plt.style.use(['presplot'])
```

constants

```python
#constants
c=2.998e8 #m/s
h=4.136e-15 #eV s
k=8.617e-5 #eV/K
q=1.602e-19 #C
t=300
omega=pi

t_s=5800
omega_s=6.8e-5
```

blackbody spectrum

```python
e_step=0.001
e=np.arange(e_step,10,e_step)
bbc=2*omega/(h**3*c**2)*e**2/(np.exp(e/(k*t))-1)
bbc_e=2*q*omega/(h**3*c**2)*e**3/(np.exp(e/(k*t))-1)
bbs_e=q*2*omega_s/(h**3*c**2)*e**3/(np.exp(e/(k*t_s))-1) #multiply by q for eV to J (answer in W)
```

plot

```python
plt.plot(e,bbs_e)
plt.xlabel("Energy (eV)")
plt.ylabel("Spectral Irradiance (W $m^{-2}$ $eV^{-1}$)")
plt.xlim([0,8])
```

calculate Rrad given a step function absorption

```python
eg=1.35
ni=1.39e8 #cm^-3

#eg=3
#ni=1.6e-6 #cm^-3

a=np.zeros(np.size(e))
a[e>eg]=1e3 #m^-1
n=4
integrand=bbc*a*4*pi*n**2
r=trapz(integrand,e)*1e-6 #1/m^3 to 1/cm^3
br=r/ni**2 #cm^3/s
```
